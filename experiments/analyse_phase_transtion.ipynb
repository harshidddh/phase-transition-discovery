```python
import numpy as np
from ml.pca_analysis import perform_pca
from ml.clustering import perform_kmeans
from ml.visualization import plot_pca_clusters, plot_observability
```

```python
# Load the data generated by run_simulation.py
X = np.load("../data/X_configs.npy")
temps = np.load("../data/y_temps.npy")
```

```python
# Run the Unsupervised ML Pipeline
X_pca, pca_model = perform_pca(X, n_components=2)
cluster_labels = perform_kmeans(X_pca, n_clusters=3)
```

```python
# Visualize Symmetry Breaking
plot_pca_clusters(X_pca, cluster_labels, save_path="../figures/pca_clusters.png")
```

```python
# Physics Validation: Prove the AI found the Order Parameter
magnetizations = np.abs(np.mean(X, axis=1))
plot_observability(temps, X_pca[:, 0], magnetizations, save_path="../figures/observability.png")
```
